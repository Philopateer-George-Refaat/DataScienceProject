## **Part 2: Data Preparation**

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import chi2
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv('fifa_player_performance_market_value.csv')
print('Dataset loaded successfully!')
print(f'Shape: {df.shape}')
```

# 2a – Meaningful Filtering Condition

```python
# --- TASK (a): Filtering ---
# We filter out players who played fewer than 5 matches.
# Reason: Players with very few matches have unreliable performance stats
# (e.g., goals, assists, minutes_played). Including them would introduce
# noise and make the analysis less meaningful for real-world decisions.

print(f'Rows BEFORE filtering: {len(df)}')

df_filtered = df[df['matches_played'] >= 5].copy()

print(f'Rows AFTER filtering (matches_played >= 5): {len(df_filtered)}')
print(f'Rows removed: {len(df) - len(df_filtered)}')
print()
print('Reason: Players with fewer than 5 matches have statistically unreliable performance data.')
```

# 2b – Encode Categorical Variables

```python
# --- TASK (b): Encoding Categorical Variables ---
# We use three encoding techniques:
#   1. Label Encoding   -> injury_prone (binary: Yes/No -> 1/0)
#   2. Ordinal Encoding -> transfer_risk_level (ordered: Low/Medium/High -> 0/1/2)
#   3. One-Hot Encoding -> position (nominal: no natural order between GK, ST, CB, etc.)

df_encoded = df_filtered.copy()

# 1. Label Encode: injury_prone
le = LabelEncoder()
df_encoded['injury_prone_encoded'] = le.fit_transform(df_encoded['injury_prone'])
print('injury_prone encoding mapping:', dict(zip(le.classes_, le.transform(le.classes_))))

# 2. Ordinal Encode: transfer_risk_level
risk_order = {'Low': 0, 'Medium': 1, 'High': 2}
df_encoded['transfer_risk_encoded'] = df_encoded['transfer_risk_level'].map(risk_order)
print('transfer_risk_level encoding mapping:', risk_order)

# 3. One-Hot Encode: position
df_encoded = pd.get_dummies(df_encoded, columns=['position'], prefix='pos', drop_first=False)

print()
print('Columns after encoding:')
print(list(df_encoded.columns))
```

# 2c – Normalize Numerical Features with StandardScaler

```python
# --- TASK (c): Normalize Numerical Features ---
# StandardScaler transforms each feature to have mean=0 and std=1.
# This is important before applying ML algorithms sensitive to scale
# (e.g., SVM, KNN, Logistic Regression).

numerical_cols = ['age', 'overall_rating', 'potential_rating', 'matches_played',
                  'goals', 'assists', 'minutes_played', 'market_value_million_eur',
                  'contract_years_left']

scaler = StandardScaler()
df_scaled = df_encoded.copy()
df_scaled[numerical_cols] = scaler.fit_transform(df_scaled[numerical_cols])

print('Numerical features after StandardScaler (mean ≈ 0, std ≈ 1):')
print(df_scaled[numerical_cols].describe().round(3))
```

# 2d – Bin a Numerical Feature into 5 Equal-Width Bins

```python
# --- TASK (d): Binning ---
# We bin 'overall_rating' from the ORIGINAL (non-scaled) data into 5 equal-width bins
# to group players by skill tier. Useful for categorical analysis or tree-based models.

df_bins = df_filtered.copy()

df_bins['rating_bin'] = pd.cut(
    df_bins['overall_rating'],
    bins=5,
    labels=['Very Low', 'Low', 'Medium', 'High', 'Very High']
)

print('Bin distribution for overall_rating (5 equal-width bins):')
bin_counts = df_bins['rating_bin'].value_counts().sort_index()
print(bin_counts)
print()
print('Bin percentages:')
print((bin_counts / len(df_bins) * 100).round(2).astype(str) + '%')

# Visualize
plt.figure(figsize=(8, 4))
bin_counts.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Distribution of Players by Overall Rating Bin')
plt.xlabel('Rating Bin')
plt.ylabel('Number of Players')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 2e – Handle Missing Values (Median / Mode)

```python

# --- TASK (e): Handle Missing Values ---
# Strategy:
#   - NUMERICAL columns -> fill with MEDIAN (robust to outliers; football stats are right-skewed)
#   - CATEGORICAL columns -> fill with MODE (most frequent value is the safest substitute)

num_cols = ['age', 'overall_rating', 'potential_rating', 'matches_played',
            'goals', 'assists', 'minutes_played', 'market_value_million_eur',
            'contract_years_left']

cat_cols = ['nationality', 'club', 'position', 'injury_prone', 'transfer_risk_level']

print('Missing values per column (before handling):')
print(df_filtered.isnull().sum())
print()

df_clean = df_filtered.copy()

# Fill numerical missing values with MEDIAN
for col in num_cols:
    missing = df_clean[col].isnull().sum()
    if missing > 0:
        median_val = df_clean[col].median()
        df_clean[col] = df_clean[col].fillna(median_val)
        print(f'Filled {col} ({missing} missing) with median = {median_val:.2f}')

# Fill categorical missing values with MODE
for col in cat_cols:
    missing = df_clean[col].isnull().sum()
    if missing > 0:
        mode_val = df_clean[col].mode()[0]
        df_clean[col] = df_clean[col].fillna(mode_val)
        print(f'Filled {col} ({missing} missing) with mode = {mode_val}')

print('Dataset is already clean — no missing values found!')
print(f'Total missing values after handling: {df_clean.isnull().sum().sum()}')
```

# 2f – Feature Selection

```python
# --- TASK (f) Part 1: Correlation Analysis (numerical features vs target) ---
# Target: market_value_million_eur
# We compute the absolute correlation of each numerical feature with the target.

corr_matrix = df_filtered[num_cols].corr()
corr_target = corr_matrix['market_value_million_eur'].drop('market_value_million_eur')
corr_target_sorted = corr_target.abs().sort_values(ascending=False)

print('=== Absolute Correlation with Target (market_value_million_eur) ===')
print(corr_target_sorted.round(4))

# Visualize
plt.figure(figsize=(9, 5))
corr_target_sorted.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Correlation of Numerical Features with Market Value')
plt.xlabel('Feature')
plt.ylabel('|Correlation|')
plt.axhline(y=0.05, color='red', linestyle='--', label='Threshold = 0.05')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# --- TASK (f) Part 2: Chi-Square Test (categorical features vs target class) ---
# Chi-square tests whether each categorical feature is statistically related to the target.
# We bin market_value into 3 classes (Low/Medium/High) to use as the target.

df_chi = df_filtered.copy()

# Bin target into 3 classes
df_chi['value_class'] = pd.cut(
    df_chi['market_value_million_eur'],
    bins=3, labels=['Low', 'Medium', 'High']
)

# Encode categorical features for chi2
cat_features_chi = ['injury_prone', 'transfer_risk_level', 'position']
df_chi_enc = df_chi[cat_features_chi].copy()
for col in cat_features_chi:
    df_chi_enc[col] = LabelEncoder().fit_transform(df_chi_enc[col])

y_chi = LabelEncoder().fit_transform(df_chi['value_class'])

chi2_scores, p_values = chi2(df_chi_enc, y_chi)

chi2_results = pd.DataFrame({
    'Feature': cat_features_chi,
    'Chi2 Score': chi2_scores.round(4),
    'P-Value': p_values.round(4)
}).sort_values('Chi2 Score', ascending=False)

print('=== Chi-Square Test (Categorical Features vs Market Value Class) ===')
print(chi2_results.to_string(index=False))
print()
print('Significant features (p < 0.05):', list(chi2_results[chi2_results['P-Value'] < 0.05]['Feature']))
```

# 2g – Remove Weak Features (with Justification)

```python
# --- TASK (g): Feature Removal with Justification ---
#
# Removed features and reasons:
# 1. player_id          -> Unique row ID, zero predictive value
# 2. player_name        -> Free text, not usable for ML without NLP
# 3. nationality        -> Very high cardinality, weak direct link to market value
# 4. club               -> Very high cardinality, would cause overfitting
# 5. contract_years_left -> Lowest correlation with market_value (seen in task f)

features_to_remove = ['player_id', 'player_name', 'nationality', 'club', 'contract_years_left']

df_selected = df_filtered.drop(columns=features_to_remove)

print('Features REMOVED:')
for feat in features_to_remove:
    print(f'  - {feat}')

print()
print('Remaining features:')
print(list(df_selected.columns))
print()
print(f'Shape before removal: {df_filtered.shape}')
print(f'Shape after  removal: {df_selected.shape}')
```

## 2h – Class Imbalance & Stratified Sampling

```python
# --- TASK (h): Class Imbalance & Stratified Sampling ---
# Class label used: transfer_risk_level (Low / Medium / High)
# We choose Stratified Sampling because it guarantees all classes
# remain proportionally represented in the sample — critical for imbalanced data.

print('=== Class Distribution BEFORE Sampling ===')
before = df_selected['transfer_risk_level'].value_counts()
print(before)
print()
print('Class percentages:')
print((before / len(df_selected) * 100).round(2).astype(str) + '%')
print()

# Stratified Sampling: keep 50% of data while preserving class ratios
df_sampled, _ = train_test_split(
    df_selected,
    test_size=0.50,
    stratify=df_selected['transfer_risk_level'],
    random_state=42
)

print('=== Class Distribution AFTER Stratified Sampling (50%) ===')
after = df_sampled['transfer_risk_level'].value_counts()
print(after)
print()
print('Class percentages:')
print((after / len(df_sampled) * 100).round(2).astype(str) + '%')
```

# 2i – Compare Class Distribution Before & After Sampling

```python
# --- TASK (i): Compare Class Distribution Before vs After Sampling ---

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
colors = ['#4C72B0', '#DD8452', '#55A868']

# Before sampling
axes[0].bar(before.index, before.values, color=colors, edgecolor='black')
axes[0].set_title('Before Sampling', fontsize=14)
axes[0].set_xlabel('Transfer Risk Level')
axes[0].set_ylabel('Count')
for i, v in enumerate(before.values):
    axes[0].text(i, v + 5, str(v), ha='center', fontweight='bold')

# After sampling
axes[1].bar(after.index, after.values, color=colors, edgecolor='black')
axes[1].set_title('After Stratified Sampling (50%)', fontsize=14)
axes[1].set_xlabel('Transfer Risk Level')
axes[1].set_ylabel('Count')
for i, v in enumerate(after.values):
    axes[1].text(i, v + 5, str(v), ha='center', fontweight='bold')

plt.suptitle('Class Distribution: Before vs After Stratified Sampling', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()

# Summary table
comparison = pd.DataFrame({
    'Before Count': before,
    'Before %': (before / len(df_selected) * 100).round(2),
    'After Count': after,
    'After %': (after / len(df_sampled) * 100).round(2)
})
print('Summary Comparison Table:')
print(comparison)
print()
print('Conclusion: Stratified sampling perfectly preserved class proportions.')
```

# **Part 4: Insight Discovery (Analytical Thinking)**

a) One statistically supported insight related to the target variable.

```python
import seaborn as sb
import matplotlib.pyplot as pylt
from scipy.stats import kruskal
```

```python
df['assist_quartiles'] = pd.qcut(df['assists'], 4,
                                 labels=['Q1 (Low)', 'Q2 (Mid-Low)', 'Q3 (Mid-High)', 'Q4 (High)'])
evidence = df.groupby('assist_quartiles')['market_value_million_eur'].mean()
print("--- Numerical Evidence: Mean Market Value by Assist Tier ---")
print(evidence)
```

```python
groups = [df[df['assist_quartiles'] == q]['market_value_million_eur'] for q in df['assist_quartiles'].unique()]
stat, p_val = kruskal(*groups) # (*) in this code unpackes the group array to pass the kruskal test to the elements in group separately

print(f"\n--- Statistical Support ---")
print(f"Kruskal-Wallis H-statistic: {stat:.4f}")
print(f"P-value: {p_val:.4e}")
```

```python
pylt.figure(figsize=(10, 6))
sb.boxplot(x='assist_quartiles', y='market_value_million_eur', data=df, palette='Spectral')

pylt.title('Part (a): Impact of Creative Output (Assists) on Market Valuation', fontsize=14)
pylt.xlabel('Player Assist Tier (Quartiles)', fontsize=12)
pylt.ylabel('Market Value (Million EUR)', fontsize=12)
pylt.grid(axis='y', linestyle='--', alpha=0.7)

pylt.show()
```

**Explanation:**
*While total minutes_played, Age, overall_rating, and potential_rating showed no statistical significance, the number of assists emerged as the most reliable indicator of a player's market worth even though the correlation between assists and market value is near zero*

*This discovery is non-trivial because it shows that "Value" in this dataset is driven by creative utility and offensive contribution rather than just pure talent ratings or physical age.*



b) One interaction-based insight involving at least two features

```python

df['age_group'] = pd.cut(df['age'], bins=[0, 23, 30, 45], labels=['Young', 'Prime', 'Veteran']) #.cuts here cuts the bins and then assign every bin to a label from the label array


df['rating_tier'] = pd.cut(df['overall_rating'], bins=[0, 75, 85, 100], labels=['Standard', 'Elite', 'World Class'])

interaction_table = df.pivot_table(index='age_group', columns='rating_tier',
                                   values='market_value_million_eur', aggfunc='mean')  # interaction_table is a grid that calculates the average market value for every specific combination of age and rating
print("--- Mean Market Value Interaction Table ---")
print(interaction_table)
```

```python
pylt.figure(figsize=(10, 6))
sb.heatmap(interaction_table, annot=True, fmt=".1f", cmap="YlGnBu", cbar_kws={'label': 'Mean Market Value (€M)'})

pylt.title('Part (b): Interaction of Age and Rating on Market Value', fontsize=14)
pylt.ylabel('Age Category')
pylt.xlabel('Performance Tier (Overall Rating)')

pylt.show()
```

**Explanation:**
*The market values "World Class" Veterans nearly equally to "Standard" Prospects, even though neither Age nor Overall Rating by themselves demonstrated statistical significance.
Typically young players should  be far more expensive because of their potential, but this interaction shows that the model values stability now as much as growth in the future.This demonstrates that a transfer is only truly "expensive" when it hits the intersection of high rating and young age*
